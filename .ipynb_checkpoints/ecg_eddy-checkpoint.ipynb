```python
# import pandas as pd# needed in order to get the data from the table 
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import pandas as pd# needed in order to get the data from the table 
filename = 'LL_2021030701.txt'
table = pd.read_csv(filename, sep=',',skiprows=(4),header=(0))
Fs = 200
Q = 30.0  # Quality factor for Notch Filter
f0 = 50.0  # Frequency to be removed from signal (Hz)
b, a = signal.iirnotch(f0, Q, Fs)#creates notch filter
# table = pd.read_csv('EM_20210306.txt', delimiter = ',')
# print(table)
# v = table.iloc[0:,1]
v = table.iloc[500:19000,1]
vFiltered = signal.lfilter(b, a, v)
t = np.arange(0,len(v)/Fs,1/Fs)
#print(table.iloc[:,0])
```

```python
from detecta import detect_peaks#needed for detect peaks
indices = detect_peaks(vFiltered, mph=-80, mpd=10)
#plotting
selectTimeRange = np.arange(500,1000)
plt.plot(t,v,'k')
plt.xlabel('time (s)')
plt.show()
plt.plot(t[selectTimeRange],v.iloc[selectTimeRange],'k')
plt.xlabel('time (s)')
plt.show()
plt.plot(t[selectTimeRange],vFiltered[selectTimeRange],'k')
plt.xlabel('time (s)')
plt.title('notch filtered')
plt.savefig(filename[:len(filename)-4]+'example')
plt.show()


plt.plot(t,vFiltered,'k')
plt.plot(t[indices],vFiltered[indices],'ro')
plt.xlabel('time (s)')
plt.show()

# print('Select two corners of zoom, middle mouse button to finish')
# pts = plt.ginput(2, timeout=-1)
peakMatrix = np.zeros((len(indices),200))
for i in range (1,len(indices)):
    peakMatrix[i,:] = np.array(vFiltered[indices[i]-100:indices[i]+100])
peakAverage = np.median(peakMatrix,axis = 0)
timePeakAverage = np.arange(-len(peakAverage)/2/Fs,len(peakAverage)/2/Fs,1/Fs)
plt.plot(timePeakAverage,peakAverage,'k')
plt.xlabel('time from R (s)')
plt.title(filename[:len(filename)-4]+' averaged')
plt.savefig(filename[:len(filename)-4])
plt.show()

```

```python
#bandpass filter
from scipy.signal import butter, lfilter

def butter_bandpass(lowcut, highcut, Fs, order=5):
    nyq = 0.5 * Fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a


def butter_bandpass_filter(data, lowcut, highcut, Fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, Fs, order=order)
    y = lfilter(b, a, data)
    return y


def run():
    import numpy as np
    import matplotlib.pyplot as plt
    from scipy.signal import freqz

    # Sample rate and desired cutoff frequencies (in Hz).
    Fs = 200.0
    lowcut = 5
    highcut = 50.0

    # Plot the frequency response for a few different orders.
    plt.figure(1)
    plt.clf()
    for order in [3, 6, 9]:
        b, a = butter_bandpass(lowcut, highcut, Fs, order=order)
        w, h = freqz(b, a, worN=2000)
        plt.plot((Fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

    plt.plot([0, 0.5 * Fs], [np.sqrt(0.5), np.sqrt(0.5)],
             '--', label='sqrt(0.5)')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Gain')
    plt.grid(True)
    plt.legend(loc='best')

    plt.show()
    vFilteredBandpass = butter_bandpass_filter(vFiltered, lowcut, highcut, Fs, order=5)
    plt.plot(vFilteredBandpass[1600:1800])
    plt.show()
    from detecta import detect_peaks#needed for detect peaks
    indices = detect_peaks(vFilteredBandpass, mph=150, mpd=10)
    plt.plot(vFilteredBandpass)
    plt.plot(t[indices],vFilteredBandpass[indices],'ro')
    plt.show()

    # print('Select two corners of zoom, middle mouse button to finish')
    # pts = plt.ginput(2, timeout=-1)
    peakMatrix = np.zeros((len(indices),400))
    for i in range (1,len(indices)-10):
        peakMatrix[i,:] = np.array(vFilteredBandpass[indices[i]-200:indices[i]+200])
    peakAverage = np.mean(peakMatrix,axis = 0)
    timePeakAverage = np.arange(-len(peakAverage)/2/Fs,len(peakAverage)/2/Fs,1/Fs)
    print(np.shape(timePeakAverage))
    print(np.shape(peakAverage))
    plt.plot(timePeakAverage,peakAverage)
    plt.xlabel('time from R (s)')
    plt.title(filename[:len(filename)-4]+' averaged')
    plt.savefig(filename[:len(filename)-4])
    plt.show()

run()    



```
